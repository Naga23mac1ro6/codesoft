```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Drop 'Date' column for visualization if needed
columns = Tesla_Prediction.drop('Date', axis=1).columns

n_cols = 2
n_rows = int(np.ceil(len(columns) / n_cols))

# Create subplots
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(9, n_rows * 3))
axes = axes.flatten()

# Plot histograms and display summary statistics
for i, column in enumerate(columns):
    ax = axes[i]
    col_data = Tesla_Prediction[column]
    ax.hist(col_data, bins=20, edgecolor='k', alpha=0.7)
    ax.set_title(f'Histogram of {column}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

    # Summary statistics text
    summary_text = Tesla_Prediction[column].describe().apply(lambda x: f'{x:.2f}').to_string()
    ax.text(0.95, 0.95, summary_text, transform=ax.transAxes, fontsize=8,
            verticalalignment='top', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# Remove any unused subplots
for ax in axes[len(columns):]:
    fig.delaxes(ax)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Drop non-numeric columns if needed (e.g., 'Date')
numeric_cols = Tesla_Prediction.drop(['Date'], axis=1)

# Set the style
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(nrows=len(numeric_cols.columns), ncols=1, figsize=(8, len(numeric_cols.columns) * 2))

# Plot individual boxplots with improved aesthetics
for i, column in enumerate(numeric_cols.columns):
    sns.boxplot(data=numeric_cols, x=column, ax=axes[i], palette="pastel")
    axes[i].set_title(f'Boxplot of {column}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel('')  # Remove the x-label
    axes[i].grid(True, linestyle='--', linewidth=0.5)

# Tight layout to reduce space between plots
plt.tight_layout()
plt.show()
```

```python
# Define a function to detect outliers using the IQR method
def find_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))]
    return outliers

# Find outliers for each numeric column
numeric_cols = Tesla_Prediction.select_dtypes(include=[np.number])
outliers_dict = {column: find_outliers_iqr(numeric_cols[column]) for column in numeric_cols.columns}

# Print the outliers
for column, outliers in outliers_dict.items():
    print(f'Outliers in {column}:')
    print(outliers)
    print('Number of outliers:', outliers.shape[0])
    print()
```

```python
import plotly.graph_objs as go
import pandas as pd

# Group by year and calculate mean of 'Open' column
yearly_data_open = Tesla_Prediction.groupby("Year")['Open'].mean().reset_index()

# Create a trace for yearly opening prices
trace_open = go.Scatter(
    x=yearly_data_open['Year'],
    y=yearly_data_open['Open'],
    mode='markers+lines',  # Connect points with lines
    name='Yearly Opening Price',
    marker=dict(size=10, color='blue', symbol='circle'),
    line=dict(color='blue', width=2)  # Line style
)

# Prepare the layout for opening prices
layout_open = go.Layout(
    title='Tesla Yearly Opening Prices',
    xaxis=dict(title='Year'),
    yaxis=dict(title='Opening Price'),
    hovermode='closest'  # Enable hover information
)

# Create the figure for opening prices
fig_open = go.Figure(data=[trace_open], layout=layout_open)

# Add annotations for each point
annotations = []
for i, row in yearly_data_open.iterrows():
    annotations.append(dict(
        x=row['Year'],
        y=row['Open'],
        text=f'Year: {row["Year"]}<br>Open: {row["Open"]:.2f}',
        showarrow=True,
        arrowhead=2,
        ax=20, ay=-30  # Position of the annotation
    ))
fig_open.update_layout(annotations=annotations)

# Display the interactive plot for opening prices
fig_open.show()
```

```python
import plotly.graph_objs as go
import pandas as pd

# Group by year and calculate mean of 'Close' column
yearly_data = Tesla_Prediction.groupby("Year")['Close'].mean().reset_index()

# Create a trace for the scatter plot of yearly closing prices
trace = go.Scatter(
    x=yearly_data['Year'],
    y=yearly_data['Close'],
    mode='markers+lines',  # Connect points with lines
    name='Yearly Closing Price',
    marker=dict(size=10, color='green', symbol='circle'),
    line=dict(color='green', width=2)  # Line style
)

# Prepare the layout
layout = go.Layout(
    title='Tesla Yearly Closing Price',
    xaxis=dict(title='Year'),
    yaxis=dict(title='Closing Price'),
    hovermode='closest'  # Enable hover information
)

# Create the figure
fig = go.Figure(data=[trace], layout=layout)

# Add annotations for each point
annotations = []
for i, row in yearly_data.iterrows():
    annotations.append(dict(
        x=row['Year'],
        y=row['Close'],
        text=f'Year: {row["Year"]}<br>Close: {row["Close"]:.2f}',
        showarrow=True,
        arrowhead=2,
        ax=20, ay=-30  # Position of the annotation
    ))
fig.update_layout(annotations=annotations)

# Display the interactive plot
fig.show()

# Save the plot as an image for inclusion in LaTeX
fig.write_image("tesla_yearly_closing_prices.png")  # Save as PNG
```

<div style="background-color: #EAE6F5; color: #4B0082; font-family: 'Arial', sans-serif; font-size: 14px; padding: 10px; border: 1px solid #FFFFFF; border-radius: 10px; margin-bottom: 10px; text-align: left; box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1);">
     <strong> What Are They?</strong><br>
    <ul style="padding-left: 20px;">
        <li>Lagged variables are features that represent past values of a particular metric. For example, yesterday’s closing price, or the closing price two days ago. These variables help models understand how past values affect future outcomes.</li>
    </ul>
    <strong>Observations:</strong><br>
    <ul style="padding-left: 20px;">
        <li><strong style="color: #4B0082;">Close Lag 1:</strong> This represents the lagged or previous closing price of the asset. "Lag 1" means it is the closing price from the previous day or the previous time period. In other words, it is the value of "Close" shifted or lagged by one time period.</li>
        <li><strong style="color: #4B0082;">Close Lag 2: </strong> Similarly, this represents the closing price that is lagged by two time periods relative to the current period. It is the closing price from two periods ago.</li>
    </ul>
</div>

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import plotly.graph_objects as go

# Sort by date (just to be sure)
Tesla_Prediction.sort_values('Date', inplace=True)

# Define features (X) and target (y)
X = Tesla_Prediction[['Open', 'High', 'Low', 'Volume', 'Adj Close']]
y = Tesla_Prediction['Close']

# Train-test split for time series data
train_size = int(len(X) * 0.8)  # 80% train, 20% test
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Function to calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# XGBoost Model
xgb_model = xgb.XGBRegressor()
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)

# Evaluate XGBoost model
mae_xgb = mean_absolute_error(y_test, xgb_pred)
rmse_xgb = np.sqrt(mean_squared_error(y_test, xgb_pred))
mape_xgb = mean_absolute_percentage_error(y_test, xgb_pred)
r2_xgb = r2_score(y_test, xgb_pred)

print("XGBoost Performance:")
print(f"MAE: {mae_xgb}")
print(f"RMSE: {rmse_xgb}")
print(f"MAPE: {mape_xgb}%")
print(f"R^2 Score: {r2_xgb}")

# Random Forest Model
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

# Evaluate Random Forest model
mae_rf = mean_absolute_error(y_test, rf_pred)
rmse_rf = np.sqrt(mean_squared_error(y_test, rf_pred))
mape_rf = mean_absolute_percentage_error(y_test, rf_pred)
r2_rf = r2_score(y_test, rf_pred)

print("\nRandom Forest Performance:")
print(f"MAE: {mae_rf}")
print(f"RMSE: {rmse_rf}")
print(f"MAPE: {mape_rf}%")
print(f"R^2 Score: {r2_rf}")

# Interactive plot with Plotly
fig = go.Figure()

# Actual stock prices
fig.add_trace(go.Scatter(x=Tesla_Prediction['Date'][train_size:], y=y_test,
                         mode='lines',
                         name='Actual',
                         line=dict(color='blue', width=2)))

# XGBoost predicted prices
fig.add_trace(go.Scatter(x=Tesla_Prediction['Date'][train_size:], y=xgb_pred,
                         mode='lines',
                         name='XGBoost Predicted',
                         line=dict(color='green', width=2)))

# Random Forest predicted prices
fig.add_trace(go.Scatter(x=Tesla_Prediction['Date'][train_size:], y=rf_pred,
                         mode='lines',
                         name='Random Forest Predicted',
                         line=dict(color='red', width=2)))

# Customize layout
fig.update_layout(title='Tesla Stock Price Prediction',
                  xaxis_title='Date',
                  yaxis_title='Stock Price (Close)',
                  legend=dict(x=0, y=1, traceorder='normal'),
                  template='plotly_dark')

# Show interactive plot
fig.show()
```

     <strong> XGBoost or Random Forest for Stock Price Prediction with Cross-Validation:</strong><br>
    <ul style="padding-left: 20px;">
        <li> The target variable (y) is Close, representing the closing price of Tesla stock. The features (X) include 'Open', 'High', 'Low', 'Volume', and 'Adj Close'.</li>
        <li> I also use cross validation splitting with XGBoost and Random Forest to check the difference in results.  Howoever both Random Forest and XGBoost performance have gone bad after using cross validation. </li>
<li> Overall Random Forest has slightly better performace than XGBoost.</li>
    </ul>

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import plotly.graph_objects as go

# Sort by date
Tesla_Prediction.sort_values('Date', inplace=True)

# Define features (X) and target (y)
X = Tesla_Prediction[['Open', 'High', 'Low', 'Volume', 'Adj Close']]
y = Tesla_Prediction['Close']

# Cross-validation strategy (K-fold with shuffle)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Function to calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Function to evaluate model performance using cross-validation
def evaluate_model(model, X, y):
    mae_scores = []
    rmse_scores = []
    mape_scores = []
    r2_scores = []

    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        mae = mean_absolute_error(y_test, y_pred)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mape = mean_absolute_percentage_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        mae_scores.append(mae)
        rmse_scores.append(rmse)
        mape_scores.append(mape)
        r2_scores.append(r2)

    return np.mean(mae_scores), np.mean(rmse_scores), np.mean(mape_scores), np.mean(r2_scores)

# XGBoost Model with Cross-Validation
xgb_model = xgb.XGBRegressor()
xgb_mae, xgb_rmse, xgb_mape, xgb_r2 = evaluate_model(xgb_model, X, y)

print("XGBoost Performance with Cross-Validation:")
print(f"Average MAE: {xgb_mae}")
print(f"Average RMSE: {xgb_rmse}")
print(f"Average MAPE: {xgb_mape}%")
print(f"Average R^2 Score: {xgb_r2}")

# Random Forest Model with Cross-Validation
rf_model = RandomForestRegressor()
rf_mae, rf_rmse, rf_mape, rf_r2 = evaluate_model(rf_model, X, y)

print("\nRandom Forest Performance with Cross-Validation:")
print(f"Average MAE: {rf_mae}")
print(f"Average RMSE: {rf_rmse}")
print(f"Average MAPE: {rf_mape}%")
print(f"Average R^2 Score: {rf_r2}")

# Visualize predictions using Plotly
fig = go.Figure()

# XGBoost predicted prices (cross-validation)
xgb_preds_cv = cross_val_predict(xgb_model, X, y, cv=kf)
fig.add_trace(go.Scatter(x=Tesla_Prediction['Date'], y=y,
                         mode='lines',
                         name='Actual',
                         line=dict(color='blue', width=2)))

fig.add_trace(go.Scatter(x=Tesla_Prediction['Date'], y=xgb_preds_cv,
                         mode='lines',
                         name='XGBoost Predicted (CV)',
                         line=dict(color='green', width=2)))

# Random Forest predicted prices (cross-validation)
rf_preds_cv = cross_val_predict(rf_model, X, y, cv=kf)
fig.add_trace(go.Scatter(x=Tesla_Prediction['Date'], y=rf_preds_cv,
                         mode='lines',
                         name='Random Forest Predicted (CV)',
                         line=dict(color='red', width=2)))

# Customize layout
fig.update_layout(title='Tesla Stock Price Prediction with Cross-Validation',
                  xaxis_title='Date',
                  yaxis_title='Stock Price (Close)',
                  legend=dict(x=0, y=1, traceorder='normal'),
                  template='plotly_dark')

# Show interactive plot
fig.show()
```
